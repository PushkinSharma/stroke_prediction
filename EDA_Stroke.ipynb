```python
df = pd.read_csv("C:\\Users\\Pushkin\\OneDrive\\Desktop\\proj\\healthcare-dataset-stroke-data.csv")
df.head()
```

```python
import plotly.express as px

fig = px.density_heatmap(df, x="age", y="bmi", nbinsx=30, nbinsy=30)
fig.show()
```

```python
fig = px.density_heatmap(df, x="age", y="bmi", nbinsx=10,nbinsy=10,facet_row="hypertension",text_auto=True)
fig.show()
```
